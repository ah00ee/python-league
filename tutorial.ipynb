```python
from league import LeagueAPI

lol = LeagueAPI("Your API KEY")
```

```python
summoner = lol.get_summoner("Hide on bush")
```

## 2. 챔피언 숙련도 불러오기

```python
# 모든 챔피언 숙련도 불러오기

all_champion_mastery = summoner.get_all_champion_mastery()
all_champion_mastery = [champion_mastery.__dict__ for champion_mastery in all_champion_mastery]
```

```python
import pandas as pd

df = pd.DataFrame(all_champion_mastery)

df
```

```python
# 특정 챔피언 숙련도 불러오기
print(lol.get_champion_by_id(99))

lux_mastery = summoner.get_champion_mastery_by_championId(99)
print(lux_mastery.__dict__)
```

```python
# 챔피언 숙련도 top 10 불러오기
top10 = summoner.get_top_champion_mastery(count=10)
for t in top10:
    print(t.__dict__)
```

## 3. 로테이션 챔피언 정보 불러오기


```python
champion_rotations = lol.get_champion_rotations()
```

```python
freeChampionIds, freeChampionIdsForNewPlayers = champion_rotations["freeChampionIds"], champion_rotations["freeChampionIdsForNewPlayers"]

freeChampion = [lol.get_champion_by_id(champion) for champion in freeChampionIds]
freeChampionForNewPlayers = [lol.get_champion_by_id(champion) for champion in freeChampionIdsForNewPlayers]

print("<Free Champion>\n==============")
for c in freeChampion:
    print(c)

maxLevel = champion_rotations["maxNewPlayerLevel"]
print(f"\n<Free Champion For New Players>\n(maximum level: {maxLevel})\n==============")
for c in freeChampionForNewPlayers:
    print(c)
```
